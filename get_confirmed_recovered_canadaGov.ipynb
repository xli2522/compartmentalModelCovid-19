```python
dataCat = []
out = []
d99 = []
```

```python
def dataReader(areaid):
    data = open("covid19June20Phase2-.txt", "r")
    i = 0
    for line in data:
        line = line.strip("\n") 
        line = line.split("\t") 
 
        if line[0] == "pruid":
            print("Call Sign:")
            print(line)
        else:
            i +=1
            if line[0] == str(areaid):
                dataCat.append(line)
            
            elif line[0] == str(99): #Travellers - 99
                d99.append(line)
  
            else:
                out.append(line)
    data.close()
    print(str(i) + " entries" )
    print()
    
def getDate(): 
    date = []
    dateN = []
    n = 0
    print("DATE")
    for i in dataCat:
        n = n + 1
        date.append(i[3])
        print(i[3])
        dateN.append(n)
    
    return dateN

def getConf(): 
    conf = []
    print("CONFIRMED CASES")
    for i in dataCat:
        conf.append(i[4])
        print(i[4])
    
    return conf
    
def getProb():
    prob = []
    for i in dataCat:
        prob.append(i[5])
    
    return prob

def getDeath():
    death = []
    for i in dataCat:
        death.append(i[6])
        
def getTest():
    test = []
    for i in dataCat:
        test.append(i[8])

def getReco():
    reco = []
    print("RECOVERED CASES")
    for i in dataCat:
        reco.append(i[9])
        print(i[9])
    return(reco)
    
def getRecoP():
    recop = []
    for i in dataCat:
        recop.append(i[10])

def getTestR():
    testR = []
    for i in dataCat:
        testR.append(i[11])
        
def getNumToday():
    numToday = []
    for i in dataCat:
        numToday.append(i[12])
    return numToday
        
def getPtoday():
    pToday = []
    for i in dataCat:
        pToday.append(i[13])
        
def getRtotal():
    rTotal = []
    for i in dataCat:
        rTotal.append(i[14])     
        
def getAll():
    for i in dataCat:
        print(str(i))
```

```python
def drawPair(x, y):
    
    fig, ax = plt.subplots()
    ax.plot(x, y)
    ax.set(xlabel='Date', ylabel='Counts/Percentages',
               title='Counts/Percentages vs. Date')
    ax.grid()
    fig.savefig("test.png")
    plt.show()
```
